```python
import math
def myeuclidean(a, b):
  temp = []
  both = zip(a, b)
  for x, y in both:
    value = (x-y) ** 2
    temp.append(value)
  return math.sqrt(sum(temp))
```

```python
# Now, let's do some evaluation on the test set. This is in itro
neighbours = 5
results = []
for t in test_set:
  t = t[:-1]
  distances = []
  for row in data:
    values = row[:-1]
    dist = distance.euclidean(t, values)
    distances.append( (dist, row)   )
  sorted_distances = sorted(distances)
  closest = sorted_distances[:neighbours]
  closestsalaries = [x[1][-1] for x in closest]
  results.append(mean(closestsalaries))
```

```python
print(results)
```

```python
truth = [x[-1] for x in test_set]
print(truth)
```

```python
# Let's compare these through plotting
import matplotlib.pyplot as plt
plt.plot(truth, color="red", label="truth")
plt.plot(results, color="blue", label="predicted")
plt.legend()
plt.grid()
plt.title("Comparing truth and predicted regression values")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(truth, results)
```

```python
# Evaluation: Mean Absolute Error
differences = []
for x, y in zip(truth, results):
  differences.append(abs(x-y))
print(differences)
print(mean(differences))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(truth, results)
```

```python
# Mean Sqaured Error: The main advantage is that it penalises large errors
# It's not intuitive
differences2 = []
for x, y in zip(truth, results):
  sqerror = (x-y) ** 2
  differences2.append(sqerror)
print("Squared Differences:", differences2)
MSE = sum(differences2) /len(truth)
print("MSE:", MSE)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(truth, results)
```

```python
# root mean square error (RMSE)
from math import sqrt
sqrt(MSE)
```
